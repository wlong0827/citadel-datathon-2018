```python
_311_service_requests_head = _311_service_requests.sample(n=300000)
_311_service_requests_head['_date'] =  pd.to_datetime(_311_service_requests_head['created_date'])
```

```python
categories = ['Water System', 'Dirty Conditions', 'Sanitation Condition', 'Rodent', 'Food Establishment', 'Air Quality', 'Indoor Air Quality', 'Food Poisoning', 'Hazardous Materials', 'Asbestos', 'Smoking', 'Drinking', 'Water Quality']

for category in categories:
    _311_service_requests_head[category] = (_311_service_requests_head['complaint_type']== category).astype('int')
```

```python
_311_service_requests_head
```

```python
_311_service_requests_head_bin = _311_service_requests_head.resample('M', on='_date').sum().dropna()
_311_service_requests_head_bin_features = _311_service_requests_head_bin[categories].copy()
_311_service_requests_head_bin_features
```
